```python
df['job'] = df['job'].str.split(',').str[0]
df['job'] = df['job'].str.split('/').str[0]
```

```python
def remove_brackets(text):
    return text.split(' (')[0]
df['job'] = df['job'].apply(remove_brackets)
```

```python
df['merchant'].unique()
```

```python
df['merchant'] = df['merchant'].str.replace('^fraud_', '', regex=True)
df['merchant'] = df['merchant'].str.replace('-','')
df['merchant'] = df['merchant'].str.replace(',','')
df['merchant'] = df['merchant'].str.replace('and','')
```

```python
df['merchant'].unique()
```

```python
df.head()
```

```python
def add_spaces(text):
    result = ''
    for char in text:
        if char.isupper() and result != '' and result[-1].islower():
            result += ' ' + char
        else:
            result += char
    return result
df['merchant'] = df['merchant'].apply(add_spaces)
```
